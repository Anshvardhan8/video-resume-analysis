# Loading Data..

```python
import pandas as pd

speech = pd.read_csv('2.csv')
emotiono = pd.read_csv('emotion.csv')
gaze = pd.read_csv('gaze.csv')
```

# Analyzing the Speech Dataset


## Descriptive Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt

emotions = speech.columns[10:-1]
avg = speech[emotions].mean()

top3 = avg.sort_values(ascending = False).head(3)
print('The Three most Dominant emotions are:')
for emo , val in top3.items():
    print(f"{emo}: {val: .2f}")
    

plt.figure(figsize = (10 , 7))
plt.pie(avg , labels = avg.index , autopct ='%1.1f%%' , startangle = 90)
plt.title('Emotion Distribution in the Speech')
plt.axis('equal')
plt.show()
```

We can see the Three most dominant emotions depicted by the person.

## Trend Analysis

```python

columns = speech.columns[10:]
for column in columns :
    plt.figure(figsize = (5 , 3.5))
    plt.bar(speech['start'] , speech[column] , width = (speech['end'] - speech['start']))
    
    plt.title(f'{column} over Time')
    plt.xlabel('Time')
    plt.ylabel(column)
    plt.grid(axis = 'y')
    plt.tight_layout()
    plt.show()
```

INFERENCES BASED ON THE ABOVE GRAPHS:

1. As we can see from the pie chart , the most dominant emotion of Person 2 is positive which is again certified by the above trend analysis , the person sounds overall positive with an exception where he talks about his two latest internships and the experience that he gained.

2. The Person sounds confindent throughout the speech except where he talks about his leading skills , and his work in Hyderabad as a framework developer.

3. There is also a dip in confidence when he talks about his two internships and how he has learned various skills through them.

4. There are some notable peaks in the hesitant graph as he talks about his work in boutique investment bank and also when he interned in Kabadi Techno.

5. The person is mostly concise except he discusses his ability of being a good team player and being a team leader. He is also repetetive when he talks about his works in different internships.

6. The person sounds very enthusiastic when he introduces himself and also while talking about he got a chance to discuss about startups , he also is very enthusiastic when he concludes .


## Correlation Analysis

```python
import seaborn as sns
selected = speech.columns[10:-1]
df = speech[selected]
corr = df.corr()

plt.figure(figsize = (6 , 4))
sns.heatmap(corr , annot = True , cmap = 'coolwarm' , vmin = -1 , vmax = 1)
plt.title("Correlation Analysis")
plt.show()
```

This heat map gives the correlation between emotions where strong correalations are generally (> 0.75 or < -0.75) and these strong correlations can be used to make inferences

# Analyzing the Emotion Dataset

```python
sns.countplot(data = emotiono , x = 'dominant_emotion')
plt.title('Frequency of Dominant Emotion')
plt.show()

emotion_counts = emotiono['dominant_emotion'].value_counts()
max_emotion = emotion_counts.idxmax()
max_count = emotion_counts.max()

print(f"The most dominant emotion is '{max_emotion}' with a count of {max_count}.")
```

```python
emotion_columns = emotiono.columns[2:-1]
fig, axes = plt.subplots(nrows=len(emotion_columns), figsize=(5 , 10), sharex=True)

x_range = range(len(emotiono))

for ax, emotion_col in zip(axes, emotion_columns):
    ax.plot(x_range, emotiono[emotion_col], label=emotion_col)
    ax.set_ylabel(emotion_col)
    ax.grid(True)
    ax.set_title(emotion_col.capitalize())

plt.xlabel('Frame Number')
plt.tight_layout()
plt.show()
```

From the graph's insights, consistent fluctuations in the emotions 'happy', 'neutral', and 'sad' emerge. These patterns suggest the speaker frequently transitions between sharing objective details, demonstrating moments of contentment or pride in their achievements, and expressing introspective reflections or challenges faced during their journey. This dynamic emotional display indicates a candidate who intertwines factual information with genuine personal experiences.

## Analyzing the Gaze Dataset


```python
print(f"Percentage of frames with eye contact: {gaze['gaze'].mean() * 100:.2f}%")
print(f"Percentage of frames with blinks: {gaze['blink'].mean() * 100:.2f}%")
```

Given that the speaker maintains eye contact in 60.2% of the frames, it indicates a commendable level of assurance and connection with the viewers. Nevertheless, there are instances where the speaker's gaze shifts, which could be attributed to moments of introspection, emphasizing particular topics, or referencing visual aids.

```python
plt.figure(figsize=(12,3))
plt.plot(gaze['image_seq'], gaze['eye_offset'], label='Eye Offset', color='green')
plt.xlabel('Frame Sequence')
plt.ylabel('Eye Offset Value')
plt.title('Eye Offset Analysis')
plt.legend()
plt.show()
```

The frequent peaks in the eye_offset graph indicate that the speaker's gaze deviates from the center regularly. This could be due to reasons like referencing notes, emphasizing certain points, or potential distractions. Such variations might also indicate moments of contemplation or trying to recall specific details or in some cases lack of confidence.
