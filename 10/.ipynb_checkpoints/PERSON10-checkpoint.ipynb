# Loading Data..

```python
import pandas as pd

speech = pd.read_csv('10.csv')
emotiono = pd.read_csv('emotion.csv')
gaze = pd.read_csv('gaze.csv')
```

# Analyzing the Speech Dataset


## Descriptive Statistics

```python
import pandas as pd
import matplotlib.pyplot as plt

emotions = speech.columns[10:-1]
avg = speech[emotions].mean()

top3 = avg.sort_values(ascending = False).head(3)
print('The Three most Dominant emotions are:')
for emo , val in top3.items():
    print(f"{emo}: {val: .2f}")
    

plt.figure(figsize = (10 , 7))
plt.pie(avg , labels = avg.index , autopct ='%1.1f%%' , startangle = 90)
plt.title('Emotion Distribution in the Speech')
plt.axis('equal')
plt.show()
```

We can see the Three most dominant emotions depicted by the person.

## Trend Analysis

```python

columns = speech.columns[10:]
for column in columns :
    plt.figure(figsize = (5 , 3.5))
    plt.bar(speech['start'] , speech[column] , width = (speech['end'] - speech['start']))
    
    plt.title(f'{column} over Time')
    plt.xlabel('Time')
    plt.ylabel(column)
    plt.grid(axis = 'y')
    plt.tight_layout()
    plt.show()
```

INFERENCES BASED ON THE ABOVE GRAPHS:

1. The values of Positive tones are exceptionally high for person10 except for a noticeable dip when they talk about their tenure as Class Representative.

2. The person10 is very confident throughout his speech but we see a drop when they talk about their work in Bad Scouts and their interest in reading books of fiction and non-fiction .

3. We see high values of hesistant through most of the speech.

4. The values of conciseness decreases as the person moves forward with the speech.

5. The person seems very enthsiastic while introducing themselves and while discussing about their role as captian in Students Committee but overall less enthusiasm is observed.



## Correlation Analysis

```python
import seaborn as sns
selected = speech.columns[10:-1]
df = speech[selected]
corr = df.corr()

plt.figure(figsize = (6 , 4))
sns.heatmap(corr , annot = True , cmap = 'coolwarm' , vmin = -1 , vmax = 1)
plt.title("Correlation Analysis")
plt.show()
```

This heat map gives the correlation between emotions where strong correalations are generally (> 0.75 or < -0.75) and these strong correlations can be used to make inferences

# Analyzing the Emotion Dataset

```python
sns.countplot(data = emotiono , x = 'dominant_emotion')
plt.title('Frequency of Dominant Emotion')
plt.show()

emotion_counts = emotiono['dominant_emotion'].value_counts()
max_emotion = emotion_counts.idxmax()
max_count = emotion_counts.max()

print(f"The most dominant emotion is '{max_emotion}' with a count of {max_count}.")
```

```python
emotion_columns = emotiono.columns[2:-1]
fig, axes = plt.subplots(nrows=len(emotion_columns), figsize=(5 , 10), sharex=True)

x_range = range(len(emotiono))

for ax, emotion_col in zip(axes, emotion_columns):
    ax.plot(x_range, emotiono[emotion_col], label=emotion_col)
    ax.set_ylabel(emotion_col)
    ax.grid(True)
    ax.set_title(emotion_col.capitalize())

plt.xlabel('Frame Number')
plt.tight_layout()
plt.show()
```

In the emotional graphs of the candidate's speech, dominant spikes in 'sad' and 'fear' are evident, with 'neutral' being less pronounced. This suggests a candidate who frequently conveys feelings of apprehension or reflects on challenges and struggles, interspersed with periods of objective communication. The pattern indicates a communicative style that's more emotionally charged, often rooted in personal experiences or concerns

## Analyzing the Gaze Dataset


```python
print(f"Percentage of frames with eye contact: {gaze['gaze'].mean() * 100:.2f}%")
print(f"Percentage of frames with blinks: {gaze['blink'].mean() * 100:.2f}%")
```

Given that the speaker maintains eye contact in 73.33% of the frames, they convey a solid level of confidence and engagement with the audience. The blink rate of 16.67% aligns reasonably well with natural behavior, suggesting the speaker is both present in the moment and actively processing their thoughts. Their performance appears balanced, merging a steady gaze with moments of introspection or reflection, making their presentation both authoritative and relatable.

```python
plt.figure(figsize=(12,3))
plt.plot(gaze['image_seq'], gaze['eye_offset'], label='Eye Offset', color='green')
plt.xlabel('Frame Sequence')
plt.ylabel('Eye Offset Value')
plt.title('Eye Offset Analysis')
plt.legend()
plt.show()
```

The frequent peaks in the eye_offset graph indicate that the speaker's gaze deviates from the center regularly. This could be due to reasons like referencing notes, emphasizing certain points, or potential distractions. Such variations might also indicate moments of contemplation or trying to recall specific details or in some cases lack of confidence.

# Overall Summary

Person10 consistently exudes confidence, but this wanes when discussing their tenure as Class Representative and involvement in Bad Scouts. While their introduction is enthusiastic, there's an overall decline in enthusiasm and conciseness as the speech progresses. With 'sad' being the dominant emotion, the speech often leans towards introspective reflections or challenges. Their 73.33% eye contact and natural 16.67% blink rate suggest a blend of confidence and genuine engagement. However, regular gaze deviations hint at occasional reliance on notes or moments of contemplation, perhaps showing a balance between preparation and authenticity.
